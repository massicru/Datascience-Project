```python
# Si crei un dataset da analizzare escludendo il 2025
df_analisi = df[df['Anno'] <= 2024].copy()
```

```python
paesi_analizzati = (
    df_analisi['Paese']
    .dropna()
    .drop_duplicates()
    .sort_values()
)

print(f"Numero di paesi analizzati: {len(paesi_analizzati)}\n")
print(paesi_analizzati.tolist())
```

```python
# Fattore di emissione del gas usato dagli autori del dataset
FATTORE_EMISSIONE_GAS = 450  # gCO₂/kWh

# Calcola la CO₂ emessa stimata per ogni Paese e anno
df_analisi['CO$_2$ Emessa Stimata (Mt)'] = (
    df_analisi['Intensità di Carbonio (gCO$_2$/kWh)']
    * df_analisi['Produzione Totale di Elettricità (TWh)']
) / 1000

# Calcola la quota di energia prodotta da fonte nucleare
df_analisi['Quota Nucleare (%)'] = (
    df_analisi['Energia Nucleare (TWh)']
    / df_analisi['Produzione Totale di Elettricità (TWh)']
) * 100

# Calcola la quota del solare sul totale della produzione rinnovabile
df_analisi['Quota Solare sul Rinnovabile (%)'] = (
    df_analisi['Energia Solare (TWh)']
    / df_analisi['Energia totale da Rinnovabili (TWh)']
) * 100

# Calcola la quota dell'eolico sul totale della produzione rinnovabile
df_analisi['Quota Eolico sul Rinnovabile (%)'] = (
    df_analisi['Energia Eolica (TWh)']
    / df_analisi['Energia totale da Rinnovabili (TWh)']
) * 100

# Calcola la quota dell'idroelettrico sul totale della produzione rinnovabile
df_analisi['Quota Idroelettrico sul Rinnovabile (%)'] = (
    df_analisi['Energia Idroelettrica (TWh)']
    / df_analisi['Energia totale da Rinnovabili (TWh)']
) * 100

# Calcola la produzione da altre fonti rinnovabili
df_analisi['Altre Fonti Rinnovabili (TWh)'] = (
    df_analisi['Energia totale da Rinnovabili (TWh)']
    - (
        df_analisi['Energia Solare (TWh)']
        + df_analisi['Energia Eolica (TWh)']
        + df_analisi['Energia Idroelettrica (TWh)']
    )
)

# Calcola la quota delle altre fonti rinnovabili sul totale delle rinnovabili
df_analisi['Quota Altre Fonti sul Rinnovabile (%)'] = (
    df_analisi['Altre Fonti Rinnovabili (TWh)']
    / df_analisi['Energia totale da Rinnovabili (TWh)']
) * 100

# Calcola la CO₂ evitata grazie al rinnovabile
df_analisi['CO$_2$ Evitata da Rinnovabili (Mt)'] = (
    df_analisi['Energia totale da Rinnovabili (TWh)']

    * FATTORE_EMISSIONE_GAS 
    
) / 1000


df_analisi
```

```python
# Raggruppa il DataFrame df_analisi in base all'anno

mondiale = (
    df_analisi.groupby('Anno')
       
       # Per ogni anno applica la funzione sum delle due colonne indicate
       .agg({
           'Energia da Fonti Fossili (TWh)': 'sum',    
           'Energia Solare (TWh)': 'sum',
           'Energia Eolica (TWh)': 'sum',
           'Energia Idroelettrica (TWh)': 'sum',
           'Altre Fonti Rinnovabili (TWh)': 'sum',
           'Energia totale da Rinnovabili (TWh)': 'sum',
           'Energia Nucleare (TWh)': 'sum',
           'Produzione Totale di Elettricità (TWh)': 'sum',
           'CO$_2$ Emessa Stimata (Mt)': 'sum',
           'CO$_2$ Evitata da Rinnovabili (Mt)': 'sum'
       })
       .reset_index() # Trasforma l'indice (Anno) in una normale colonna del DataFrame
)

# Si crea una nuova colonna chiamata "Quota mondiale rinnovabili (%)"
mondiale['Quota mondiale rinnovabili (%)'] = (
    # Si prende il totale dell'energia rinnovabile per ogni anno
    mondiale['Energia totale da Rinnovabili (TWh)']

    # si divide per la produzione totale di elettricità dello stesso anno
    / mondiale['Produzione Totale di Elettricità (TWh)']
    
) * 100 # Ottengo la percentuale

# Si crea una nuova colonna chiamata "Quota mondiale fonti fossili (%)"
mondiale['Quota mondiale fonti fossili (%)'] = (
    mondiale['Energia da Fonti Fossili (TWh)']
    / mondiale['Produzione Totale di Elettricità (TWh)']
    
) * 100

# Si crea una nuova colonna chiamata "Quota mondiale nucleare (%)"
mondiale['Quota mondiale nucleare (%)'] = (
    mondiale['Energia Nucleare (TWh)']
    / mondiale['Produzione Totale di Elettricità (TWh)']
) * 100

mondiale['Intensità mondiale (gCO$_2$/kWh)'] = (
    mondiale['CO$_2$ Emessa Stimata (Mt)']
    * 1000
    / mondiale['Produzione Totale di Elettricità (TWh)']
)

mondiale
```

```python
# Seleziona le righe del 2000 e del 2024
riga_2000 = mondiale[mondiale['Anno'] == 2000]
riga_2024 = mondiale[mondiale['Anno'] == 2024]

# Estrae le quote del 2000
rinnovabili_2000 = riga_2000['Quota mondiale rinnovabili (%)'].values[0]
fossili_2000 = riga_2000['Quota mondiale fonti fossili (%)'].values[0]
nucleare_2000 = riga_2000['Quota mondiale nucleare (%)'].values[0]

# Estrae le quote del 2024
rinnovabili_2024 = riga_2024['Quota mondiale rinnovabili (%)'].values[0]
fossili_2024 = riga_2024['Quota mondiale fonti fossili (%)'].values[0]
nucleare_2024 = riga_2024['Quota mondiale nucleare (%)'].values[0]

# Stampa i risultati
print("=== Quote nel 2000 ===")
print(f"Rinnovabili: {rinnovabili_2000:.2f}%")
print(f"Fonti fossili: {fossili_2000:.2f}%")
print(f"Nucleare: {nucleare_2000:.2f}%")

print("\n=== Quote nel 2024 ===")
print(f"Rinnovabili: {rinnovabili_2024:.2f}%")
print(f"Fonti fossili: {fossili_2024:.2f}%")
print(f"Nucleare: {nucleare_2024:.2f}%")

# Calcola le variazioni
print("\n=== Variazione 2000-2024 ===")
print(f"Rinnovabili: {rinnovabili_2024 - rinnovabili_2000:+.2f} punti percentuali")
print(f"Fonti fossili: {fossili_2024 - fossili_2000:+.2f} punti percentuali")
print(f"Nucleare: {nucleare_2024 - nucleare_2000:+.2f} punti percentuali")
```

```python
# Crea una lista vuota che conterrà, per ogni Paese,
# la pendenza e l'intercetta della retta di regressione
pendenze = []

# Raggruppa il DataFrame per Paese.
# Ad ogni iterazione:
# - "paese" contiene il nome del Paese
# - "gruppo" contiene tutte le righe relative a quel Paese
for paese, gruppo in df_analisi.groupby('Paese'):

    # Ordina i dati del Paese in ordine crescente di anno.
    # Non è strettamente necessario per polyfit,
    # ma rende i dati ordinati e più leggibili.
    gruppo = gruppo.sort_values('Anno')

    # Calcola la retta di regressione lineare:
    # y = mx + q
    #
    # dove:
    # x = Anno
    # y = Quota di energia rinnovabile (%)
    #
    # Restituisce:
    # m = pendenza della retta
    # q = intercetta della retta
    m, q = np.polyfit(
        gruppo['Anno'],
        gruppo['Quota Rinnovabili (%)'],
        1      # grado 1 = retta
    )

    # Salva i risultati ottenuti per il Paese corrente
    # all'interno della lista "pendenze"
    pendenze.append({
        'Paese': paese,
        'Continente': gruppo['Regione'].iloc[0],
        'Pendenza': m,
        'Intercetta': q
    })

# Trasforma la lista di dizionari in un DataFrame
df_pendenze = pd.DataFrame(pendenze)

# Ordina i Paesi in ordine decrescente di pendenza.
# I Paesi con la pendenza maggiore sono quelli che,
# mediamente, hanno aumentato più rapidamente
# la quota di energia rinnovabile nel periodo analizzato.
df_pendenze = (
    df_pendenze
    .sort_values(
        'Pendenza',
        ascending=False
    )
    .reset_index(drop=True)
)

# Visualizza il DataFrame finale
df_pendenze
```

```python
import pycountry # contiene un database di paesi, lingue e suddivisioni territoriali secondo gli standard ISO

def nome_to_iso3(paese):
    try:
        return pycountry.countries.search_fuzzy(paese)[0].alpha_3

        # pycountry.countries : Accede all'elenco di tutti i paesi presenti nella libreria.
        # .search_fuzzy(paese): Cerca il paese anche se il nome non è scritto esattamente
        # [0] : Poiché search_fuzzy restituisce una lista di risultati, [0] prende il primo risultato, che è quello più probabile.
        # .alpha_3: Prende il codice ISO a tre lettere del paese.
        
    except LookupError:
        return None

df_pendenze['ISO3'] = df_pendenze['Paese'].apply(nome_to_iso3)
```

```python
# Si verifichi che la prima scelta sia quella corretta

for nome in df_pendenze["Paese"].unique():
    print(f"\n{nome}")
    try:
        risultati = pycountry.countries.search_fuzzy(nome)
        for paese in risultati:
            print(f"  - {paese.name} ({paese.alpha_3})")
    except LookupError:
        print("  Nessuna corrispondenza trovata")
```

```python
import geopandas as gpd
import matplotlib.patheffects as pe

# Caricamento della mappa

world = gpd.read_file(
    "Natural_Earth_quick_start/110m_cultural/"
    "ne_110m_admin_0_countries.shp"
)

# Unione con df_pendenze

# Il merge viene fatto usando il nome del Paese,
# perché df_pendenze non contiene la colonna ISO3.
world = world.merge(
    df_pendenze,
    left_on="ADMIN",
    right_on="Paese",
    how="left"
)

# Preparazione dei dati geografici

# Scala cromatica simmetrica centrata sullo zero
lim = world["Pendenza"].abs().max()

# Punti interni ai Paesi per posizionare le etichette
label_points = world.geometry.representative_point()

world["label_x"] = label_points.x
world["label_y"] = label_points.y

# Creazione della figura

fig = plt.figure(figsize=(20, 11))


# Mappa mondiale principale

ax = fig.add_axes([0.03, 0.08, 0.72, 0.82])

world.plot(
    column="Pendenza",
    cmap="BuGn",
    vmin=-lim,
    vmax=lim,
    linewidth=0.5,
    edgecolor="0.65",
    missing_kwds={
        "color": "lightgray",
        "edgecolor": "0.7"
    },
    ax=ax
)

# Mostra i numeri per tutti i Paesi tranne quelli europei
for _, row in world.iterrows():

    if (
        pd.notna(row["Pendenza"])
        and row["Continente"] != "Europe"
    ):

        ax.text(
            row["label_x"],
            row["label_y"],
            f'{row["Pendenza"]:.2f}',
            ha="center",
            va="center",
            fontsize=8,
            fontweight="bold",
            color="black",
            path_effects=[
                pe.withStroke(
                    linewidth=2.5,
                    foreground="white"
                )
            ],
            zorder=10
        )

ax.set_xlim(-180, 180)
ax.set_ylim(-60, 90)
ax.set_axis_off()

ax.set_title(
    "Variazione della quota di energia rinnovabile per Paese (2000–2024)",
    fontsize=20,
    fontweight="bold",
    pad=15
)

# Riquadro Europa ingrandito

ax_eu = fig.add_axes([0.75, 0.42, 0.24, 0.45])

world.plot(
    column="Pendenza",
    cmap="BuGn",
    vmin=-lim,
    vmax=lim,
    linewidth=0.7,
    edgecolor="0.55",
    missing_kwds={
        "color": "lightgray",
        "edgecolor": "0.7"
    },
    ax=ax_eu
)

# Limiti geografici dell'Europa
ax_eu.set_xlim(-12, 45)
ax_eu.set_ylim(34, 72)

# Mostra soltanto i Paesi con Continente == "Europe"
for _, row in world.iterrows():

    dentro_riquadro = (
        -12 <= row["label_x"] <= 45
        and 34 <= row["label_y"] <= 72
    )

    if (
        pd.notna(row["Pendenza"])
        and row["Continente"] == "Europe"
        and dentro_riquadro
    ):

        ax_eu.text(
            row["label_x"],
            row["label_y"],
            f'{row["Pendenza"]:.2f}',
            ha="center",
            va="center",
            fontsize=7,
            fontweight="bold",
            color="black",
            path_effects=[
                pe.withStroke(
                    linewidth=2.2,
                    foreground="white"
                )
            ],
            zorder=10
        )

ax_eu.set_title(
    "Dettaglio Europa",
    fontsize=13,
    fontweight="bold",
    pad=8
)

ax_eu.set_axis_off()

# Barra dei colori

sm = plt.cm.ScalarMappable(
    cmap="BuGn",
    norm=plt.Normalize(
        vmin=-lim,
        vmax=lim
    )
)

sm.set_array([])

cax = fig.add_axes([
    0.80,   # posizione orizzontale
    0.12,   # posizione verticale
    0.015,  # larghezza
    0.23    # altezza
])

cbar = fig.colorbar(
    sm,
    cax=cax
)

cbar.set_label(
    "Pendenza",
    fontsize=11,
    fontweight="bold"
)

# Visualizzazione

plt.show()
```
